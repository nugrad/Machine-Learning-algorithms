# Supervised Learning

### Decision Tree Classifier :

Decision Tree is a Supervised Machine Learning Algorithm that uses a set of rules to make decisions, similarly to how humans make decisions.

There are three methods to create decision tree :

- ID3 Algorithm (based on entropy and information gain)  
- CART Algorithm (based on gini index and gini gain)
- C4.5 Algorithm (based on gain ratio)


#### Lets study working of ID3 Algorithm so that you may know how decision tree algorithm works from scratch :

Steps in making decisiont tree using ID3 algorithm :
    
 - First of all, we need to decide root node of a tree.

 - Among all the features given, we find entropy and information gain of every feature.

 - The feature that gives lowest entropy (or highest information gain) will be choosen as root node of tree

 - Then for second level of tree, we again choose nodes with lowest entropy (or highest information gain)

 - This goes on till all nodes in leaf nodes are divided into separate classes i-e all yes on one side and all no on other. or you can say this goes on till leaf nodes get pure (separate classes)

    

This is the way to create decision tree using ID3 algorithm. 

If you want to create decision tree using CART algorithm, the only difference will be, you will be calculating gini index (instead of entropy) for every node and node with lowest gini index (or highest gini gain) will be choosen as root node and process goes on.

If you want to create decision tree using C4.5 algorithm, the only difference will be, you will be calculating gain ratio (instead of entropy or gini index) for every node and node with highest gain ratio will be choosen as root node and process goes on till leaf nodes become pure.


#### Pruning in decision tree :
Pruning is a data compression technique in machine learning and search algorithms that "reduces the size of
decision trees" by removing sections of the tree that are non-critical (not important) and redundant to classify instances.
Pruning reduces the complexity of the final classifier, and hence improves predictive accuracy


## Decision Tree using Library

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset = pd.read_csv(r"C:\Users\msohail.afzal\Desktop\lastlab\iris.data", names=names)
#dataset
array = dataset.values
X = array[:,0:4]
Y = array[:,4]
t_size = 0.20 
seed = 7 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=t_size,random_state=seed)
```

```python
# Decision Tree using Entropy (with out pruning parameter)

dt = DecisionTreeClassifier( criterion='entropy')
dt.fit(X_train, Y_train) # training the model
predictions = dt.predict(X_test) # now testing on test data to get class of test data
print((accuracy_score(Y_test, predictions))) # comparing results predicted by model with actual
```

```python
# Decision Tree using Entropy (with pruning parameter "ccp_alpha")

dt = DecisionTreeClassifier( criterion='entropy', ccp_alpha = 0.015) # ccp_alpha is pruning parameter
dt.fit(X_train, Y_train) # training the model
predictions = dt.predict(X_test) # now testing on test data to get class of test data
print((accuracy_score(Y_test, predictions))) # comparing results predicted by model with actual
```

```python
# Decision Tree using Gini (with out pruning parameter)

dt = DecisionTreeClassifier( criterion='gini')
dt.fit(X_train, Y_train) # training the model
predictions = dt.predict(X_test) # now testing on test data to get class of test data
print((accuracy_score(Y_test, predictions))) # comparing results predicted by model with actual
```

```python
# Decision Tree using Gini (with pruning parameter "ccp_alpha")

dt = DecisionTreeClassifier( criterion='gini', ccp_alpha = 0.015) # ccp_alpha is pruning parameter
dt.fit(X_train, Y_train) # training the model
predictions = dt.predict(X_test) # now testing on test data to get class of test data
print((accuracy_score(Y_test, predictions))) # comparing results predicted by model with actual
```
