## Implementing Naive Bayes from Library

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score



names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset = pd.read_csv(r"C:\Users\msohail.afzal\Desktop\lastlab\iris.data", names=names)


#dataset
array = dataset.values
X = array[:,0:4]
Y = array[:,4]
t_size = 0.20 # training size is 0.2 (20 percent), testing size will be automatically
seed = 7 #if this seed is set to any constant number then it will generate same number
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=t_size,random_state=seed)
```

```python
from sklearn.naive_bayes import GaussianNB
# Naive Bayes Algorithm
nb = GaussianNB() 
```

```python
nb.fit(X_train, Y_train) # training the model
predictions = nb.predict(X_test) # now testing on test data to get class of test data
print((accuracy_score(Y_test, predictions))) # comparing results predicted by model with actual
```
